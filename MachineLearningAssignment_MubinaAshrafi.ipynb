```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, BatchNormalization, Flatten, Dropout
from keras.utils import to_categorical
import cv2
from keras.optimizers import Adam
```

```
dataset = pd.read_csv('/content/sample_data/mnist_test.csv')
```

```
print(dataset.shape)
dataset.describe()
```

```
dataset = dataset.iloc[:,:].values
x = dataset[:, 1:785]
y = dataset[:,0]
```

```
temp = []
for i in x :
  im = i.reshape(28, 28)
  temp.append(im)
x = np.array(temp)
```

```
x = x.reshape(-1, 28, 28, 1)
y = to_categorical(y)
```

```
model = Sequential()
model.add(Conv2D(64, (5, 5), input_shape=(28, 28, 1), activation='relu', padding='same'))
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(10, activation='softmax'))
 
adam = Adam(lr=.0001)
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=adam)
model.summary()
```

```
model.fit(x, y, batch_size=200, epochs=5, validation_split=0.2)
```
